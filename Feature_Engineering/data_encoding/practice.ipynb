1. Nominal/ One Hot Encoding 

This data encoding technique will create new features based on the number of categories that we have and place 1 if it is present in this row or 0 if not present

example: dress_color field has - [Red, Greeen, Blue] then we create 3 new features Red, Green and Blue and put 1 and rest all 0's 

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({'color': ['red', 'blue', 'green', 'blue', 'red']})
df.head()
```

```python
# Create a instance of OneHotEncoder
encoder = OneHotEncoder()
```

```python
# Fit and transform the data
encoded = encoder.fit_transform(df[['color']]).toarray()
```

```python
import pandas as pd
encoder_df = pd.DataFrame(encoded, columns=encoder.get_feature_names_out(['color']))
encoder_df
```

```python
encoder.transform([['blue']]).toarray()
```

```python
pd.concat([df, encoder_df], axis=1)
```

Example Working on tips dataset

```python
import seaborn as sns
df = sns.load_dataset('tips')
df.head()
```

```python
df['day'].unique()
```

```python
encoder = OneHotEncoder()
encoded = encoder.fit_transform(df[['day']]).toarray()
```

```python
encoder.get_feature_names_out()
```

```python
encoded_df = pd.DataFrame(
    encoded,
    columns=encoder.get_feature_names_out()
)
encoded_df.head()
```

```python
df_final = pd.concat([df, encoded_df], axis=1)

df_final.head()
```

Label Encoding 

This method assigns a unique numerical values for the categorical data

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
```

```python
df = pd.DataFrame({
    'color': ['red', 'blue', 'green', 'blue', 'red']})
df.head()
```

```python
lbl_encoder.fit_transform(df[['color']])
```

```python
lbl_encoder.transform([['red']])
```

```python
lbl_encoder.transform([['blue']])
```

The problem with this method is when we assign unique labels our model may start thinking red may have a higher capacity value because red has 2 which is higher when compared to green(1) and blue(0)

so we move from nominal to ordinal data

Ordinal Data Encoding

In Ordinal Data Encoding Technique the numeric values are assigned based on its position in the order

```python
import pandas as pd
from sklearn.preprocessing import OrdinalEncoder
```

```python
df= pd.DataFrame({'size': ['small', 'medium', 'large', 'medium', 'small']})
df.head()
```

```python
ordinal_encoder = OrdinalEncoder(categories=[['small', 'medium', 'large']])
ordinal_encoded = ordinal_encoder.fit_transform(df[['size']])
ordinal_encoded
```
