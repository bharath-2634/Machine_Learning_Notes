Outlier Detection 

There are 2 techniques to detect Outliers
1. IQR Method
2. Z-score

1. Z- Score Approach


```python
import numpy as np
import matplotlib.pyplot as plt

data = np.array([10, 12, 14, 15, 18, 100])

x = np.arange(len(data))

plt.scatter(x, data)
```

```python
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std
print(np.abs(z_scores))  # print z-scores
outliers = np.abs(z_scores) > 1
plt.scatter(x[outliers], data[outliers])  # highlight

plt.title("Scatter Plot with Outlier")
plt.xlabel("Index")
plt.ylabel("Value")
plt.show()
```

```python
plt.boxplot(data)
plt.title("Boxplot for Outlier Detection")
plt.ylabel("Value")
plt.show()
```

2. IQR Approach

Formula: IQR=Q3−Q1

Find Q1 (25th percentile)
Find Q3 (75th percentile)
Compute IQR
Define bounds:

Lower Bound: Q1−1.5×IQR

Upper Bound: Q3+1.5×IQR

```python
import numpy as np

data = np.array([10, 12, 14, 15, 18, 100])

Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = data[(data < lower_bound) | (data > upper_bound)]
print("Outliers:", outliers)
```

```python
import matplotlib.pyplot as plt

plt.boxplot(data)

# Plot outliers manually (for clarity)
x_positions = [1] * len(outliers)  # boxplot is at x=1
plt.scatter(x_positions, outliers)

plt.title("Boxplot with IQR Outliers Highlighted")
plt.ylabel("Value")
plt.show()
```
